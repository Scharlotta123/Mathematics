```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-np.pi, np.pi, 201)
k=1
plt.plot(x, np.cos(1*x), 'r-', x, np.cos(-4*x), 'b--')
plt.title("Графики функции")
plt.xlabel("Переменная x")
plt.ylabel("Функция y(x)")
plt.legend(labels=["y(k,x)=cos(1∙x)", " y(k,x)=cos(-4∙x)"],loc="lower right")
plt.show()
```
